```python
def getchars(data):
#     a = [ord(x) for x in list(data)]
    aa = data + [0]
    bb = [0] + data
    c = [abs(m-n) for m,n in zip(aa,bb)]
    d = [i for i, j in enumerate(c) if j == 32]
    f = [x for x in d if not x-1 in d]
    e = [x-1 for x in f]
    rem = e + f
#     print ('rem ', rem)
    for index in sorted(rem, reverse=True):
        del aa[index]
#     print(''.join([chr(x) for x in aa]))
    ab = [x for x in aa if x != 0]
    if len(ab) < len(data):
        getchars(ab)
    else:
        print('result, ',ab, len(ab))
    return ab
    
```
